```python
import os
main_path = '/Volumes/Epileptology01/Oleg/SCN1A/R1648/'
img_name = 'Fused.tif'
list_of_names = []
for i,item in tbl['image_path'].items():
    if isinstance(item,str):
    # if len([item])>3:
        # print(item)
        # item = os.path.normpath(item)
        # print(Path(item).resolve())
        # p = os.path.abspath(os.path.join(main_path,item,img_name))
        # p = p.replace(r'\\',r'/')
        item = item.replace('\\','/')
# # Option 1: Normalize to OS-native format (e.g., backslashes on Windows)
#         native_path = str(p.resolve())  # or just str(p)
#         # Option 2: Convert to consistent POSIX-style (always forward slashes)
#         posix_path = p.as_posix()
#         print(posix_path)
#         # list_of_names.append()
        list_of_names.append(os.path.join(main_path,item,img_name))
```

```python
import numpy as np

list_of_names = list(np.unique(list_of_names))
```
